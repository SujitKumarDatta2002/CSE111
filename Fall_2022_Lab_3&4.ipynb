```python
class Shape():
    def __init__(self,shape,base,height) -> None:
        self.shape = shape
        self.base = base
        self.height = height
    def area(self):
        if self.shape.upper() == 'TRIANGLE':
            area = 0.5 * self.base * self.height
            print(f"Area: {area}")
        elif self.shape.upper() == 'SQUARE':
            area = self.base ** 2
            print(f"Area: {area}")
        elif self.shape.upper() == 'RHOMBUS':
            area = 0.5 * self.base * self.height
            print(f"Area: {area}")
        elif self.shape.upper() == 'RECTANGLE':
            area = self.base * self.height
            print(f"Area: {area}")
        else:
            print('Area: Shape unknown')
        

# Write your code here

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```
