```python
def fraction(a,b):
  if b == 0:
    return 0
  elif a == 0:
    return 0
  else:
    result = (a/b) - (a//b)
    return result

output = fraction(0,5)
print(output)
```

```python
BMI = 0
def BMI_calculator(height,weight):
  meterHeight = height / 100
  global BMI
  BMI = round(weight / (meterHeight**2),1)
  if BMI < 18.5:
    return "Underweight"
  elif 18.5<= BMI <= 24.9:
    return "Normal" 
  elif 25<= BMI <= 30:
    return "Overweight" 
  # elif BMI > 30:
  else:
    return "Obese" 

result = BMI_calculator(152,48)
print("Score is",BMI,"."+"You are",result)
```

```python
def sumOfDivisor(min,max,div):
  total = 0
  for i in range(min,max,div):
    if i % div == 0:
      total += i
  return total

result = sumOfDivisor(3,16,3)
print(result)
```

```python
def vowelFinder(name):
  vowel = "aeiouAEIOU"
  vowel_count = 0
  vowel_find = ''
  for i in name:
    if i in vowel:
      vowel_count += 1
      vowel_find += i+","
    else:
      pass
  if vowel_count != 0:
    print("Vowels",vowel_find,"Total number of vowels:",vowel_count)
  else:
    print("No vowels in the name!")

vowelFinder('Steve Jobs')
```

```python
def palindrome(word):
  string = ''
  for i in word:
    if 65<=ord(i)<=90 or 97<=ord(i)<= 122:
      string += i     
  forwardString = string[:]
  backwardString = string[-1:-len(string)-1:-1]

  if forwardString == backwardString:
    print("Palindrome")
  else:
    print("Not a palindrome")

palindrome('nurses run')
```
