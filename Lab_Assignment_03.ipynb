```python
class Calculator():
  def __init__(self):
    print("Let's Calculate!")
  def add(self, val1, val2):
    result = val1 + val2
    print(f"Value 1: {val1}\nOperator: +\nValue 2: {val2}")
    print("Result: ",result)
  def subtract(self, val1, val2):
    result = val1 - val2
    print(f"Value 1: {val1}\nOperator: -\nValue 2: {val2}")
    print("Result: ",result)
  def multiply(self, val1, val2):
    result = val1 * val2
    print(f"Value 1: {val1}\nOperator: *\nValue 2: {val2}")
    print("Result: ",result)
  def divide(self, val1, val2):
    result = val1 / val2
    print(f"Value 1: {val1}\nOperator: /\nValue 2: {val2}")
    print("Result: ",result)

c1 = Calculator()

val1 = int(input("Enter your first value: "))
operator = input("Enter operator: ")
val2 = int(input("Enter your second value: "))
if operator == "+":
  c1.add(val1,val2)
elif operator == "-":
  c1.subtract(val1,val2)
elif operator == "*":
  c1.multiply(val1,val2)
else:
  c1.divide(val1,val2)
```

```python
class Shape():
    def __init__(self,shape,base,height) -> None:
        self.shape = shape
        self.base = base
        self.height = height
    def area(self):
        if self.shape.upper() == 'TRIANGLE':
            area = 0.5 * self.base * self.height
            print(f"Area: {area}")
        elif self.shape.upper() == 'SQUARE':
            area = self.base ** 2
            print(f"Area: {area}")
        elif self.shape.upper() == 'RHOMBUS':
            area = 0.5 * self.base * self.height
            print(f"Area: {area}")
        elif self.shape.upper() == 'RECTANGLE':
            area = self.base * self.height
            print(f"Area: {area}")
        else:
            print('Area: Shape unknown')
        

# Write your code here

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
class Calculator():
    def __init__(self) -> None:
        print('Calculator is raedy!')

    def showCalculation(self):
        print(f"{self.value1} {self.sign} {self.value2} ={val} ")
    def calculate(self,value1,value2,sign):
        self.value1 = value1
        self.value2 = value2
        self.sign = sign
        if self.sign == '+':
            result = self.value1 + self.value2
            return result
        elif self.sign == '-':
            result = self.value1 - self.value2
            return result
        elif self.sign == '*':
            result = self.value1 * self.value2
            return result
        elif self.sign == '/':
            result = self.value1 / self.value2
            return result

# Write your code here

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```
