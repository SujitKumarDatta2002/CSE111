01.Write a product class with a customized constructor where product ID, product price and
product weight are used as parameters. The class has 2 functions named update price,
printdetails.
Update price function will update the price as the parameter and print details will print the
product ID, product price and weight.

```python
class Product():
    def __init__(self, product_ID, product_price, product_weight) -> None:
        self.product_ID =product_ID
        self.product_price = product_price
        self.product_weight = product_weight
    def updatePrice(self, price):
        self.product_price = price

    def printDetails(self):
        print(f"Product ID: {self.product_ID}\nProduct Price: {self.product_price}\nProduct Weight: {self.product_weight}")
```
